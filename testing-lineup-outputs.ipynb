```python
    import pandas as pd
    from starlette.testclient import TestClient
    from fasthtml.common import *
    from fasthtml.jupyter import JupyUvi, HTMX
    from dataclasses import dataclass, field
    from typing import List, Literal

date='2024-08-24'
plrs = pd.read_csv('data/players.csv')
apps = pd.read_csv('data/player_apps.csv')
subs = pd.read_csv('data/subs.csv')

df = apps.query("game_date==@date") \
.merge(plrs, on='player_id') \
.merge(subs, left_on=['game_date', 'player_id'], right_on=['game_date', 'player_on_id'], how='left', ) \
.to_dict(orient='records')
```

```python
plr_recs = df

plr_recs[11]
```

```python
app_div = [(Div(
    Div(
        Div(
            H1(int(plr_rec['shirt_no']),),
            style='display:inline-block;'
        ),
        Img(
            src=f'img/{plr_rec['player_id']}.png',
            style='display:inline-block;width: 50px;'
        ),
        style='display: inline-block;'
    ),
    Div(
        Div(
            plr_rec['player_name'],
            style='font-weight: bold;'
        ),
        Div(
            plr_rec['role'],
            style='font-weight:light;color:slategray;'
        ),
        Div(
            plr_rec['position']
        ),
    style='display: inline-block;'
    ),
    Div(
        Div(
            H1(
                f"On for {plr_rec['player_off_id']}" if plr_rec['player_on_id'] == plr_rec['player_id'] else ''
            ),
            H1(
                f'({int(plr_rec['sub_min'])})',
            ) if not pd.isna(plr_rec['sub_min']) else '', style='font-weight: bold; color: white;'),
        style='display:inline-block;'
    ),
    style=f'background:{'lightgreen' if not plr_rec['role']=='sub' else 'lightblue'}'
), Hr()) for plr_rec in plr_recs]

show(*app_div)
```
